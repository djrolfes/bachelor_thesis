```python
a = 1
Ns = np.array([i for i in range(3,20,2)])
lattices = [gen_all_points_no_normalisation(i) for i in Ns]
ks = np.array([[0,0,0], [0.05,0,0],[0.1,0,0], [0.2,0,0]])
rs = np.zeros((ks.shape[0], len(Ns)))


for index, lattice in enumerate(lattices):
    print(2*index+3)
    comm = La_Lb_commutator(lattice, a,3, ang=genLaNew)
    for ik, k in enumerate(ks):
        vec = fourier_vector(lattice, a, k)
        rs[ik, index] = calc_r(comm, vec)

```

```python
for ik, k in enumerate(ks):
    plt.plot(1/Ns, rs[ik, :], "o--", label=f"k:{k}")
plt.xlabel("mean distances")
plt.ylabel("r")
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend(loc="best")
```

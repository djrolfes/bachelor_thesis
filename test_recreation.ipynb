```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook', 'grid']) #dependency: SciencePlots
from genz_recreate import gen_all_points, gen_La_forward, gen_La
from lattice_actions import calc_mean_distance
from commutators import test_angular_momentum_comutator, fourier_vector, calc_r, La_Lb_commutator
```

i = 1, j = 0, a = 1

```python
a = 1
Ns = np.array([i for i in range(3,12)])
lattices = [gen_all_points(i) for i in Ns]
ks = np.array([[0,0,0], [0.1,0,0],[0.5,0,0], [1,0,0]])
rs = np.zeros((ks.shape[0], len(Ns)))


for index, lattice in enumerate(lattices):
    print(index)
    comm = test_angular_momentum_comutator(lattice, a, i=1, j=0, ang=gen_La_forward)
    for ik, k in enumerate(ks):
        vec = fourier_vector(lattice, a, k)
        rs[ik, index] = calc_r(comm, vec)

```

```python

mean = []
for lattice in lattices:
    mean.append(calc_mean_distance(lattice))
```

```python
for ik, k in enumerate(ks):
    plt.plot(mean, rs[ik, :], "o--", label=f"k:{k}")
plt.xlabel("mean distances")
plt.ylabel("r")
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend(loc="best")
```
